```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pilih kolom mean saja sebagai contoh agar tidak terlalu penuh
mean_cols = [c for c in df.columns if "_mean" in c]

plt.figure(figsize=(12, 10))
# Menghitung korelasi antar kolom mean
correlation_matrix = df[mean_cols].corr()

sns.heatmap(
    correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1
)
plt.title("Matriks Korelasi Fitur MFCC Mean", fontsize=14, fontweight="bold")
plt.show()
```

```python
# =====================================================
# PRINT KORELASI DI ATAS THRESHOLD
# =====================================================

threshold = 0.5

print(f"\nPasangan fitur dengan korelasi > {threshold}:\n")

for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        
        corr_value = correlation_matrix.iloc[i, j]
        
        if abs(corr_value) > threshold:
            print(
                f"{correlation_matrix.columns[i]} "
                f"<-> "
                f"{correlation_matrix.columns[j]} "
                f"= {corr_value:.2f}"
            )
```

```python
# Threshold
threshold = 0.5

# Simpan nilai korelasi
corr_values = []

print(f"\nPasangan fitur dengan korelasi > {threshold}:\n")

# Loop matrix
for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)):

        corr_value = correlation_matrix.iloc[i, j]

        # Ambil korelasi di atas threshold
        if abs(corr_value) > threshold:

            print(
                f"{correlation_matrix.columns[i]} "
                f"<-> "
                f"{correlation_matrix.columns[j]} "
                f"= {corr_value:.2f}"
            )

            corr_values.append(abs(corr_value))

# Hitung rata-rata korelasi
average_corr = sum(corr_values) / len(corr_values)

print("\n==========================")
print(f"Jumlah pasangan fitur : {len(corr_values)}")
print(f"Rata-rata korelasi    : {average_corr:.2f}")
```

```python
# =====================================================================
# RANDOM FOREST CLASSIFICATION - DATA PREPARATION
# =====================================================================
print("\n" + "="*70)
print("KLASIFIKASI GENRE MUSIK - RANDOM FOREST")
print("="*70)

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    confusion_matrix, classification_report, accuracy_score,
    precision_score, recall_score, f1_score, precision_recall_fscore_support
)
import seaborn as sns

# Persiapan fitur dan target
X = df[mean_std_cols]  # Gunakan fitur MFCC yang sudah dipilih
y = df['Genre']

print(f"\n📊 Dataset Info:")
print(f"   Shape: {X.shape}")
print(f"   Samples: {len(X)}")
print(f"   Features: {X.shape[1]}")
print(f"   Genres: {y.nunique()}")

# Encode label genre
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print(f"\n🏷️  Genre Labels:")
for idx, genre in enumerate(label_encoder.classes_):
    count = sum(y_encoded == idx)
    print(f"   {idx}: {genre:12s} ({count:3d} samples)")

# Standardisasi fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"\n✓ Fitur telah distandardisasi")

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print(f"\n📈 Train-Test Split (80-20):")
print(f"   Training: {len(X_train)} samples")
print(f"   Testing:  {len(X_test)} samples")
```

```python
# =====================================================================
# RANDOM FOREST - TRAINING & EVALUATION
# =====================================================================
print("\n" + "="*70)
print("TRAINING RANDOM FOREST CLASSIFIER")
print("="*70)

print("\n🔄 Training Random Forest...")
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)
rf_model.fit(X_train, y_train)
print("✓ Training selesai")

# Prediksi
y_pred_rf = rf_model.predict(X_test)
y_pred_train_rf = rf_model.predict(X_train)

# Evaluasi
train_acc_rf = accuracy_score(y_train, y_pred_train_rf)
test_acc_rf = accuracy_score(y_test, y_pred_rf)
prec_rf = precision_score(y_test, y_pred_rf, average='macro', zero_division=0)
rec_rf = recall_score(y_test, y_pred_rf, average='macro', zero_division=0)
f1_rf = f1_score(y_test, y_pred_rf, average='macro', zero_division=0)

print(f"\n📊 Hasil Evaluasi:")
print(f"   Training Accuracy : {train_acc_rf:.4f}")
print(f"   Test Accuracy     : {test_acc_rf:.4f}")
print(f"   Precision (macro) : {prec_rf:.4f}")
print(f"   Recall (macro)    : {rec_rf:.4f}")
print(f"   F1-Score (macro)  : {f1_rf:.4f}")

# Confusion Matrix
cm_rf = confusion_matrix(y_test, y_pred_rf)
print(f"\n🔤 Confusion Matrix:")
print(cm_rf)

# Visualisasi Confusion Matrix
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues', 
            xticklabels=label_encoder.classes_,
            yticklabels=label_encoder.classes_,
            cbar_kws={'label': 'Count'},
            ax=ax)
ax.set_title('Random Forest - Confusion Matrix', fontsize=14, fontweight='bold')
ax.set_ylabel('True Label', fontsize=12)
ax.set_xlabel('Predicted Label', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Classification Report
print(f"\n📋 Classification Report:")
print(classification_report(y_test, y_pred_rf, target_names=label_encoder.classes_))
```

```python
# =====================================================================
# PER-GENRE PERFORMANCE ANALYSIS
# =====================================================================
import numpy as np

print("\n" + "="*70)
print("PER-GENRE PERFORMANCE ANALYSIS")
print("="*70)

precision_per_class, recall_per_class, f1_per_class, _ = precision_recall_fscore_support(
    y_test, y_pred_rf
)

print("\n🎵 Performa Per-Genre:")
print("-" * 70)
print(f"{'Genre':<15} {'Precision':<12} {'Recall':<12} {'F1-Score':<12}")
print("-" * 70)

for idx, genre in enumerate(label_encoder.classes_):
    print(f"{genre:<15} {precision_per_class[idx]:<12.4f} {recall_per_class[idx]:<12.4f} {f1_per_class[idx]:<12.4f}")

print("-" * 70)

# Visualisasi performa per-genre
fig, ax = plt.subplots(figsize=(12, 6))

x = np.arange(len(label_encoder.classes_))
width = 0.25

bars1 = ax.bar(x - width, precision_per_class, width, label='Precision', color='steelblue', edgecolor='black', alpha=0.8)
bars2 = ax.bar(x, recall_per_class, width, label='Recall', color='coral', edgecolor='black', alpha=0.8)
bars3 = ax.bar(x + width, f1_per_class, width, label='F1-Score', color='lightgreen', edgecolor='black', alpha=0.8)

ax.set_xlabel('Genre', fontsize=12, fontweight='bold')
ax.set_ylabel('Score', fontsize=12, fontweight='bold')
ax.set_title('Random Forest - Performa Per-Genre', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(label_encoder.classes_, rotation=45, ha='right')
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.3)
ax.set_ylim([0, 1.1])

plt.tight_layout()
plt.show()

print("\n✓ Per-genre analysis selesai")
```

```python
# =====================================================================
# CONTOH PREDIKSI & MISCLASSIFICATION ANALYSIS
# =====================================================================
import numpy as np

print("\n" + "="*70)
print("ANALISIS PREDIKSI & MISCLASSIFICATION")
print("="*70)

# Fungsi untuk prediksi dengan confidence
def predict_genre(sample_features, model=rf_model, label_encoder=label_encoder):
    """Prediksi genre dari fitur MFCC"""
    if len(sample_features.shape) == 1:
        sample_features = sample_features.reshape(1, -1)
    
    predicted_class = model.predict(sample_features)[0]
    probabilities = model.predict_proba(sample_features)[0]
    sorted_idx = np.argsort(probabilities)[::-1]
    
    result = {
        'predicted_genre': label_encoder.classes_[predicted_class],
        'confidence': probabilities[predicted_class],
        'top_3_predictions': [
            {
                'genre': label_encoder.classes_[sorted_idx[i]],
                'probability': probabilities[sorted_idx[i]]
            }
            for i in range(min(3, len(sorted_idx)))
        ]
    }
    
    return result

# Misclassification analysis
misclassified_indices = np.where(y_pred_rf != y_test)[0]
print(f"\n📊 Total Misclassified: {len(misclassified_indices)} dari {len(y_test)} samples")
print(f"   Error Rate: {len(misclassified_indices)/len(y_test)*100:.2f}%")

# Contoh prediksi yang benar
print(f"\n✅ Contoh Prediksi BENAR (5 Sampel Pertama dari Test Set):")
print("-" * 80)

for i in range(min(5, len(X_test))):
    test_sample = X_test[i:i+1]
    true_label = label_encoder.classes_[y_test[i]]
    
    result = predict_genre(test_sample)
    
    if result['predicted_genre'] == true_label:
        print(f"\nSampel {i+1}:")
        print(f"  True Genre:       {true_label}")
        print(f"  Predicted Genre:  {result['predicted_genre']}")
        print(f"  Confidence:       {result['confidence']:.4f}")

print("\n" + "-" * 80)

# Contoh prediksi yang salah
if len(misclassified_indices) > 0:
    print(f"\n❌ Contoh Prediksi SALAH (Misclassified Samples):")
    print("-" * 80)
    
    for i in range(min(3, len(misclassified_indices))):
        idx = misclassified_indices[i]
        test_sample = X_test[idx:idx+1]
        true_label = label_encoder.classes_[y_test[idx]]
        
        result = predict_genre(test_sample)
        
        print(f"\nSampel {i+1}:")
        print(f"  True Genre:       {true_label}")
        print(f"  Predicted Genre:  {result['predicted_genre']}")
        print(f"  Confidence:       {result['confidence']:.4f}")
        print(f"  Top 3 Predictions:")
        for j, pred in enumerate(result['top_3_predictions'], 1):
            print(f"    {j}. {pred['genre']:<15} ({pred['probability']:.4f})")

print("\n" + "="*80)
print("✅ KLASIFIKASI RANDOM FOREST SELESAI!")
print("="*80)

# Summary
print(f"""
📈 RINGKASAN HASIL KLASIFIKASI:

🎵 Dataset:
   - Total Samples: {len(df)}
   - Training: {len(X_train)} | Testing: {len(X_test)}
   - Features: {X.shape[1]} | Genres: {len(label_encoder.classes_)}

🏆 Akurasi Model:
   - Training Accuracy: {train_acc_rf:.4f}
   - Test Accuracy:     {test_acc_rf:.4f}
   - Precision (macro): {prec_rf:.4f}
   - Recall (macro):    {rec_rf:.4f}
   - F1-Score (macro):  {f1_rf:.4f}

📊 Misclassification:
   - Misclassified: {len(misclassified_indices)} samples
   - Error Rate: {len(misclassified_indices)/len(y_test)*100:.2f}%

🎯 Top 3 Important Features:
   - {importance_df.iloc[0]['Feature']}: {importance_df.iloc[0]['Importance']:.4f}
   - {importance_df.iloc[1]['Feature']}: {importance_df.iloc[1]['Importance']:.4f}
   - {importance_df.iloc[2]['Feature']}: {importance_df.iloc[2]['Importance']:.4f}
""")
```
